```python
from sklearn import svm
from sklearn import datasets
clf = svm.SVC()
iris = datasets.load_iris()
X, y = iris.data, iris.target
clf.fit(X, y)
```

```python
import pickle
s = pickle.dumps(clf)
clf2 = pickle.loads(s)
clf2.predict(X[0:1])
```

pickle本身是将类序列化，序列化后需要保存至硬盘进行持久化

```python
f=open('model.txt','wb')
pickle.dump(s,f,0) 
f.close() 
```

```python
f=open('model.txt','r')  
clf2 = pickle.loads(s)
f.close()
clf2.predict(X[0:1]) [0]
```

### 使用joblib模块

```python
import joblib
joblib.dump(clf2, 'filename.pkl')
```

```python
clf = joblib.load('filename.pkl')
clf2.predict(X[0:1]) [0]
```
